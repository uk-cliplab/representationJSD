```python
%matplotlib inline
%load_ext autoreload
%autoreload 2
import numpy as np
import torch
import torch.utils.data
import matplotlib.pyplot as plt
import repitl.kernel_utils as ku
from ITL_utils import JSD_difference,QJSD
import repitl.kernel_utils as ku
```

```python
from  torch.distributions import multivariate_normal
covY = torch.tensor([[1.0,0.7],[0.7,1.0]])
import matplotlib.style
import matplotlib as mpl
mpl.style.use('bmh')
n = 1000
d = 2
rho =0.99
mean1 = [0,0]
mean2 = [0,0]
cov1 = np.identity(d)
cov2 = np.identity(d)
cov2[0,1] = rho
cov2[1,0] = rho
cov2 = cov2
cov1 = cov1*0.5
X = torch.tensor(np.random.multivariate_normal(mean1, cov1, n))
Y = torch.tensor(np.random.multivariate_normal(mean2, cov2, n))

plt.plot(X[:,0],X[:,1], '.', label = r'$\mathbf{X}$' )
plt.plot(Y[:,0],Y[:,1], '.', label = r'$\mathbf{Y}$' )
plt.legend()
```

```python
N = X.shape[0]
M = Y.shape[0]
n_perm = 20
sigmas = np.linspace(0.01,3,70) 
J = torch.zeros((len(sigmas)))
JSD = torch.zeros((len(sigmas)))
JSDperm = torch.zeros((len(sigmas),n_perm))
Z = torch.cat([X, Y], dim=0).double()

for i,sigma in enumerate(sigmas):
    for j in range(n_perm):
        Zperm = Z[torch.randperm(len(Z))]    
        Zx = Zperm[:N,:]
        Zy = Zperm[-M:,:]
        JSDperm[i,j] = QJSD(Zx,Zy, sigma = sigma)


    JSD[i] = QJSD(X,Y, sigma = sigma)
    

JSD = torch.log2(torch.exp(JSD))
JSDperm = torch.log2(torch.exp(JSDperm))
```

```python
import matplotlib.style
import matplotlib as mpl
mpl.style.use('bmh')
Ex_JSDperm = torch.mean(JSDperm, axis = 1)


J = JSD - Ex_JSDperm
best_sigma = sigmas[np.argmax(J)]
plt.plot(sigmas,JSD, label = r'$JSD(\mathbf{X},\mathbf{Y})$')
plt.plot(sigmas,Ex_JSDperm, color = 'orange', alpha = 0.5,label = r'$\mathbb{E}\left[JSD(\mathbf{Z}_x,\mathbf{Z}_y)\right]$')
plt.plot(sigmas,J, label = r'$\mathbb{J}^{\kappa_\sigma}(\mathbf{X},\mathbf{Y})$')
plt.plot(best_sigma, J[np.argmax(J)], '*', color = 'darkred',markersize = 10, label = r'$\sigma_*$')
plt.xlabel(r'$\sigma$')
plt.legend()
```
